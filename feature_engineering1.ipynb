```python
def transform_date(df):
    df['day'] =  pd.DatetimeIndex(df['click_time']).day
    df['weekday'] = pd.DatetimeIndex(df['click_time']).weekday
    df['hour'] = pd.DatetimeIndex(df['click_time']).hour
    df['minutes'] = pd.DatetimeIndex(df['click_time']).minute
    df['seconds'] = pd.DatetimeIndex(df['click_time']).second
    return df
```

```python
def count_nunique(groupby_col, n_col, newname, filename):
    ncol_count = pd.DataFrame(full_data[groupby_col + [n_col]].groupby(groupby_col)[n_col].nunique()).reset_index()
    ncol_count.rename(columns={n_col: newname}, inplace = True)
    ncol_count.to_csv(filename, index = False)
    return ncol_count
```

```python
napp_byip = count_nunique(['ip'], 'app', 'napp_byip', 'napp_byip.csv')
```
